```python
import cv2
import numpy as np
import os


# Function to process a single image
def process_image(image_path, output_dir):
    # Read the input image
    image = cv2.imread(image_path)
    
    # Convert the image to grayscale for thresholding
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Thresholding to segment regions of interest
    _, binary_mask = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

    # Apply morphological operations to improve the mask
    kernel = np.ones((5, 5), np.uint8)
    binary_mask = cv2.morphologyEx(binary_mask, cv2.MORPH_CLOSE, kernel)

    # Label the segmented regions
    _, labeled_mask = cv2.connectedComponents(binary_mask)

    # Save the generated mask
    mask_filename = os.path.join(output_dir, os.path.basename(image_path).replace('.JPG', '_mask.JPG'))
    cv2.imwrite(mask_filename, labeled_mask)

# Input and output directories
input_dir = r'C:\Users\shanmuga\OneDrive\Documents\mini project\mini project\UDD\UDD\UDD5\train\train_image' # Adjust path as needed
output_dir = r'C:\Users\shanmuga\OneDrive\Documents\mini project\mini project\UDD\UDD\UDD5\train\train_mask' # Adjust path as needed

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Process each image in the input directory
for image_file in os.listdir(input_dir):
    if image_file.endswith('.JPG'):
        image_path = os.path.join(input_dir, image_file)
        process_image(image_path, output_dir)
```
